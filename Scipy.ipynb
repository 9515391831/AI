#Integration functions

*General integration*

1.The quad function calculates the integral of a function which has one variable.

```
from scipy import integrate
```

```
help(integrate.quad)
```

```
g=sc.integrate.quad(lambda x:special.exp2(x),0,1)
g
```

*Double integration*

1.The dblquad function calculates double integral of a function which has two variables.

```
h=lambda x,y:x+y
i=lambda x:1
j=lambda x:-1
k=integrate.dblquad(h,0,1,i,j)
k
```

```
help(integrate.dblquad)
```

```
sc.info(integrate)
```

```
sc.source(integrate)
```

#Linear algebra
