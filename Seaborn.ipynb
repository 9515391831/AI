```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
a=sns.load_dataset("tips")
a
```

```python
b=sns.load_dataset("flights")
b
```

```python
sns.relplot(x="passengers",y="month",data=b)
```

```python
sns.relplot(x="passengers",y="month",hue="year",data=b)
```

```python
sns.relplot(x="time",y="tip",data=a,kind="line")
```

```python
sns.catplot(x="time",y="tip",data=a)
```

```python
sns.catplot(x="day",y="total_bill",data=a)#this is the scatter plot which is default for catplot
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=a)#box plot
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=a)#violin plot
```

```python
sns.catplot(x="day",y="total_bill",kind="boxen",data=a)#boken plot
```

# Multi plots

```python
a=sns.load_dataset("iris")
a
```

```python
b=sns.FacetGrid(a,col="species")
b.map(plt.hist,"sepal_length")
```

```python
c=sns.PairGrid(a)
c.map(plt.scatter)
```

```python
sns.set(style='darkgrid')
c=sns.PairGrid(a)
c.map(plt.scatter)
```

```python
sns.set(style='dark')
c=sns.PairGrid(a)
c.map(plt.scatter)
```

```python
c=sns.color_palette()
sns.palplot(c)
```
